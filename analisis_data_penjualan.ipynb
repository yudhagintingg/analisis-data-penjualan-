```python

df['Tanggal_Pemesanan'] = pd.to_datetime(df['Tanggal_Pemesanan'])
df['Bulan'] = df['Tanggal_Pemesanan'].dt.to_period('M').astype(str)

# Agregasi omzet per bulan
monthly_sales = df.groupby('Bulan')['Total_Penjualan'].sum().reset_index()
monthly_sales['Bulan'] = pd.to_datetime(monthly_sales['Bulan'])

# Model regresi
X = monthly_sales.index.values.reshape(-1, 1)
y = monthly_sales['Total_Penjualan']

model = LinearRegression()
model.fit(X, y)

# Prediksi bulan depan
next_month = [[len(monthly_sales)]]
predicted_sales = model.predict(next_month)
print(f"Prediksi penjualan bulan depan: Rp{predicted_sales[0]:,.0f}")
```

```python
# Hitung total belanja tiap customer
customer_df = df.groupby('Customer_ID',)['Total_Penjualan'].sum().reset_index()

# Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
customer_df['Cluster'] = kmeans.fit_predict(customer_df[['Total_Penjualan']])

print(customer_df.head())
```

```python
# Buat matriks customer vs produk
pivot_df = df.pivot_table(index='Customer_ID', columns='Nama_Produk', values='Jumlah', fill_value=0)

from sklearn.metrics.pairwise import cosine_similarity

# Cari similarity antar user
similarity = cosine_similarity(pivot_df)
similarity_df = pd.DataFrame(similarity, index=pivot_df.index, columns=pivot_df.index)

# Contoh rekomendasi untuk CUST1096
similar_customers = similarity_df['CUST1096'].sort_values(ascending=False)[1:]
print("Customer mirip:", similar_customers.head())

# Produk dari customer mirip
target_user = 'CUST1096'
similar_user = similar_customers.index[0]
recommend_products = pivot_df.loc[similar_user][pivot_df.loc[target_user]==0]
recommend_products = recommend_products[recommend_products > 0]
print("Rekomendasi produk:", recommend_products.index.tolist())
```
